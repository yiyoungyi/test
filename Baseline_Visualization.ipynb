```python
#혼잡을 red
colordict = {0 : 'lightblue', 1 : 'lightgreen', 2 : 'orange', 3 : 'red'}
```

### create a base map

```python
folium.Map
```

```python
#서울 중심
latitude = 37.5312271
longitude = 126.9500134
fine_dust_map = folium.Map(location=[latitude, longitude], zoom_start=12,tiles='stamentoner')
fine_dust_map
```

```python
for lat, lon, dust_q, pm10, city in zip(df['X좌표(WGS)'], 
                                            df['Y좌표(WGS)'], 
                                            df['dust_quartile'], 
                                            df['PM-10'], 
                                            df['전철역명']):
    folium.CircleMarker(
        [lat, lon],
        radius=.2*pm10,
        popup = folium.Popup(city + '<br>' 
                 +'PM-10 score: ' + str(pm10) + '<br>', max_width=300
                ),
        color='b',
        
        fill_color=colordict[dust_q],
        fill=True,
        fill_opacity=0.7
        ).add_to(fine_dust_map)
fine_dust_map
```
